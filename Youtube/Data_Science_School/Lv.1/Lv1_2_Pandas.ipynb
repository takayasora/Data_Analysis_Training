# 第四部：データの結合

```python
df_person1 = pd.read_csv("personNo_1.csv",index_col="社員番号")
df_person2 = pd.read_csv("personNo_2.csv",index_col="社員番号")
```

```python
df_person1
```

```python
df_person2
```

```python
#左外部結合（Left Join）
df_person1.join(df_person2)
```

```python
#右外部結合（Right Join）
df_person1.join(df_person2,how="right")
```

```python
#内部結合（Inner Join）
df_person1.join(df_person2,how="inner")
```

```python
#外部結合（Outer Join）
df_person1.join(df_person2,how="outer")
```

## 2.margeによる結合

```python
df_person = df_person1.join(df_person2,how="outer")
df_section = pd.read_csv("sectionNo.csv",index_col="組織番号")
```

```python
df_person
```

```python
df_section
```

```python
#キーは組織番号で結合
pd.merge(df_person,df_section,on="組織番号")
```

```python
#キーを複数設定
pd.merge(df_person,df_section,on=["組織番号","組織名"])
```

```python
#キーを複数設定
pd.merge(df_person,df_section,on=["組織番号","組織名"],how="outer")
```
