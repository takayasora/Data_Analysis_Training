```python
#Pythonã§ä¾¡æ ¼å¸¯ã‚’åˆ†é¡ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã®ã‚³ãƒ¼ãƒ‰ä¾‹ã‚’ç”Ÿæˆã—ã¦ã¿ã¾ã™ã€‚ ã¾ãšã€å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆã—ã¾ã™ã€‚

# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã®ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import tree
#æ¬¡ã«ã€å­¦ç¿’ç”¨ãƒ‡ãƒ¼ã‚¿ã¨è©•ä¾¡ç”¨ãƒ‡ãƒ¼ã‚¿ã‚’èª­ã¿è¾¼ã¿ã¾ã™ã€‚

# ãƒ‡ãƒ¼ã‚¿ã®èª­ã¿è¾¼ã¿
train = pd.read_csv("./original/train.csv")
test = pd.read_csv("./original/test.csv")
#ãƒ‡ãƒ¼ã‚¿ã®å½¢å¼ã‚’ç¢ºèªã—ã¾ã™ã€‚

# ãƒ‡ãƒ¼ã‚¿ã®å½¢å¼ã‚’ç¢ºèª
print(train.shape) # (2000, 22)
print(test.shape) # (1000, 21)
#å­¦ç¿’ç”¨ãƒ‡ãƒ¼ã‚¿ã«ã¯ã€ä¾¡æ ¼å¸¯ã‚’è¡¨ã™price_rangeã¨ã„ã†ã‚«ãƒ©ãƒ ãŒã‚ã‚Šã¾ã™ãŒã€è©•ä¾¡ç”¨ãƒ‡ãƒ¼ã‚¿ã«ã¯ã‚ã‚Šã¾ã›ã‚“ã€‚ ã“ã‚ŒãŒç›®çš„å¤‰æ•°ã§ã™ã€‚ ç‰¹å¾´é‡ã¨ç›®çš„å¤‰æ•°ã«åˆ†ã‘ã¾ã™ã€‚

# ç‰¹å¾´é‡ã¨ç›®çš„å¤‰æ•°ã«åˆ†ã‘ã‚‹
X_train = train.drop("price_range", axis=1)
y_train = train["price_range"]
X_test = test.copy()
#ç‰¹å¾´é‡ã®ã‚¹ã‚±ãƒ¼ãƒªãƒ³ã‚°ã‚’è¡Œã„ã¾ã™ã€‚

# ç‰¹å¾´é‡ã®ã‚¹ã‚±ãƒ¼ãƒªãƒ³ã‚°
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
#ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›å¸°ãƒ¢ãƒ‡ãƒ«ã‚’ä½œæˆã—ã€å­¦ç¿’ç”¨ãƒ‡ãƒ¼ã‚¿ã§å­¦ç¿’ã•ã›ã¾ã™ã€‚

# ãƒ­ã‚¸ã‚¹ãƒ†ã‚£ãƒƒã‚¯å›å¸°ãƒ¢ãƒ‡ãƒ«ã®ä½œæˆã¨å­¦ç¿’
model = tree.DecisionTreeClassifier(max_depth=3,random_state=1)
model.fit(X_train, y_train)
#å­¦ç¿’ã—ãŸãƒ¢ãƒ‡ãƒ«ã§è©•ä¾¡ç”¨ãƒ‡ãƒ¼ã‚¿ã®ä¾¡æ ¼å¸¯ã‚’äºˆæ¸¬ã—ã¾ã™ã€‚

# è©•ä¾¡ç”¨ãƒ‡ãƒ¼ã‚¿ã®ä¾¡æ ¼å¸¯ã®äºˆæ¸¬
y_pred = model.predict(X_test)
#äºˆæ¸¬çµæœã‚’csvãƒ•ã‚¡ã‚¤ãƒ«ã«å‡ºåŠ›ã—ã¾ã™ã€‚

# äºˆæ¸¬çµæœã®å‡ºåŠ›
output = pd.DataFrame({"index": test["id"], "price_range": y_pred})
output.to_csv("./submission/submission_dt_v2.csv", index=False,header=False)
#ä»¥ä¸ŠãŒã€Pythonã§ä¾¡æ ¼å¸¯ã‚’åˆ†é¡ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã®ã‚³ãƒ¼ãƒ‰ä¾‹ã§ã™ã€‚123 ã“ã®ã‚³ãƒ¼ãƒ‰ã¯ã‚ãã¾ã§å‚è€ƒã§ã™ã€‚å®Ÿéš›ã«å‹•ä½œã•ã›ã‚‹ã«ã¯ã€train.csvã¨test.csvã¨ã„ã†ãƒ•ã‚¡ã‚¤ãƒ«ãŒå¿…è¦ã§ã™ã€‚ã¾ãŸã€ãƒ¢ãƒ‡ãƒ«ã‚„ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿ã¯ä»–ã«ã‚‚è‰²ã€…ã‚ã‚Šã¾ã™ã®ã§ã€è‡ªåˆ†ã§è©¦ã—ã¦ã¿ã¦ãã ã•ã„ã€‚ğŸ˜Š

#å—ä¿¡ã—ãŸãƒ¡ãƒƒã‚»ãƒ¼ã‚¸. Pythonã§ä¾¡æ ¼å¸¯ã‚’åˆ†é¡ã™ã‚‹ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã®ã‚³ãƒ¼ãƒ‰
```
